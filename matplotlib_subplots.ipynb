```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[1].plot(a, b)
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x, y)
axes[0][1].plot(a, b)
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1)
for ax in axes:
    ax.plot(x, y)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x, y)
axes[0][1].plot(x, y)
axes[1][0].plot(x, y)
axes[1][1].plot(x, y)

fig.subplots_adjust(left=None,
    bottom=None,
    right=None,
    top=None,
    wspace=0.9,
    hspace=0.3,)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

fig.savefig("Subplots.png",bbox_inches="tight")

plt.show()
```
