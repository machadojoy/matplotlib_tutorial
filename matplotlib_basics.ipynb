```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0, 10)
```

```python
y = 2 * x
```

```python
plt.plot(x, y)
plt.title("A title")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
plt.plot(x, y)
plt.title("With limits")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.xlim(0, 6)
plt.ylim(0, 15)
```
