```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')


axes2 = fig.add_axes([0.2, 0.6, 0.25, 0.25])
axes2.plot(x, y)
axes2.set_title('Small Figure')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')
axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)

axes2 = fig.add_axes([0.2, 0.6, 0.25, 0.25])
axes2.plot(a, b)
axes2.set_title('Zoomed in')
axes2.set_xlim(1, 2)
axes2.set_ylim(0, 100)
```

```python
fig = plt.figure(figsize=(2,2),dpi=100)
axes1 = fig.add_axes([0, 0, 1, 1]) # Full figure
axes2 = fig.add_axes([1, 1, 0.25, 0.25]) # Starts at top right corner!

axes1.plot(a, b)

fig.savefig('new_figure.png',bbox_inches='tight')
```
